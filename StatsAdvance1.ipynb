Question1 : Define the z-statistic and explain its relationship to the standard normal distribution. How is the z-statistic used in hypothesis testing?

Answer: The z-statistic is a number representing how many standard deviations above or below the mean population the score derived from a z-test. It is calculated by subtracting the population or, sample mean from the individual data point and then dividing by the population or sample standard deviation. It is closely related to the standard normal distribution, which is a theoretical distribution with a mean of 0 and a standard deviation of 1. When a data point is standardized into a z-score, it can be compared to the standard normal distribution to determine its relative position within the population or sample.

The z-statistic is used in hypothesis testing to assess whether a sample mean is significantly different from a hypothesized population mean. The null hypothesis, which states that there is no difference between the sample mean and the population mean, is assumed to be true until there is sufficient evidence to reject it.

The formula to calculate the z-statistics : z = (x - μ) / σ
where:
x is the individual data point
μ is the population or sample mean
σ is the population or sample standard deviation

Once the z-statistic is calculated, it can be compared to the critical values of the standard normal distribution. If the absolute value of the z-statistic is greater than the critical value, then the null hypothesis can be rejected at the specified level of significance.

Let’s take an example:
We need to know whether the average height of students in a class is significantly different from the national average height of 170 cm. we take a sample of 30 students and find that the sample mean height is 175 cm with a standard deviation of 10 cm. The standard error of the mean is 10 cm / sqrt(30) = 1.83 cm.

The z-statistic is calculated as (175 - 170) / 1.83 = 2.73. If we choose a level of significance of 0.05, the critical values are -1.96 and 1.96. Since 2.73 is greater than 1.96, we can reject the null hypothesis and conclude that the average height of students in the class is significantly different from the national average height.

The value of a binomial is obtained by multiplying the number of independent trials by the successes.

The binomial distribution can be used under the following conditions
•	n, the number of trials is finite and independent
•	each trial results in two mutually exclusive and exhaustive
•	the probability of success (p) must remain the same for all trials.

The binomial distribution is closely related to the Bernoulli distribution. Here's how they are connected:

Bernoulli Distribution: This is the simplest discrete probability distribution, representing a single trial with two possible outcomes: success (coded as 1) and failure (coded as 0). The probability of success is denoted by  p, and the probability of failure is  1-p. The Bernoulli distribution can be represented as:
P(X=k)=pk (1 - p)1−k, k∈{0,1}

Binomial Distribution: This distribution generalizes the Bernoulli distribution to n independent trials (each with the same probability p of success). The binomial distribution counts the number of successes in these n trials. It is defined as:

P(X=k)= (nk) pk(1−p)n−k, k = 0,1,2,…,n
Where (nk)   is the binomial coefficient, representing the number of ways to choose k successes in  n trials.

In the conclusion we can say that, the Bernoulli distribution is a special case of the binomial distribution where,  n = 1. The binomial distribution can be thought of as the sum of n independent Bernoulli trials. This relationship highlights how the binomial distribution builds upon the foundation laid by the Bernoulli distribution.


Question5: What are the key properties of the Poisson distribution, and when is it appropriate to use this distribution?

Answer: A Poisson distribution is known to be the probability distribution that results from a Poisson experiment. As in the binomial distribution, we will not know the number of trials, or the probability of success on a certain trail. The average number of successes (wins) will be given for a certain time interval. The average number of successes is known as “Lambda” and denoted by the symbol λ.

Key Properties of a Poisson Distribution:

•	Discrete: It deals with discrete random variables, meaning the outcomes are countable (e.g., the number of phone calls in an hour, the number of bacteria in a culture).
•	Mean and Variance: The mean (λ) and variance of a Poisson distribution are equal. This means that the average number of events in a given interval is also a measure of the variability of those events.
•	No Upper Limit: The Poisson distribution can take on any non-negative integer value, meaning there's no theoretical maximum number of events that can occur.
•	Independence: Events occur independently of each other. The occurrence of one event does not affect the probability of another event occurring.
•	Rare Events: The Poisson distribution is often used to model rare events. However, it can also be used to model more frequent events as long as the average rate of occurrence is constant.

The Poisson distribution is appropriate to use when:

•	Events are rare: The distribution is particularly useful for modeling events that occur infrequently.
•	Events occur independently: The occurrence of one event does not affect the probability of another event occurring.
•	Events occur at a constant rate: The average rate of occurrences (λ) remains constant over the time or space interval being considered.
•	The time or space interval is fixed: The distribution is used for a specific, fixed interval.


Question6: Define the terms "probability distribution" and "probability density function" (PDF). How does a PDF differ from a probability mass function (PMF)?

Answer: A probability distribution is a statistical function that describes all the possible values and likelihoods that a random variable can take within a given range. This range will be bounded between the minimum and maximum possible values. However, where the possible value is likely to be plotted on the probability distribution depends on several factors. These factors include the distribution’s mean (average), standard deviation, skewness, and kurtosis.

The probability density function (PDF) is a statistical expression that defines the probability that some outcome will occur. In this function, the probability is the percentage of a dataset's distribution that falls between two criteria. PDF is used by financial analysts to understand how returns are distributed in order to evaluate the risk and expectations of investment prices and returns.

Probability Mass Function (PMF) and Probability Density Function (PDF) are both mathematical concepts used in probability theory to describe the probability distributions of random variables. However, they have some fundamental differences, as explained below:

Definition:
PMF: A PMF is used to describe the probability distribution of a discrete random variable. It assigns probabilities to each possible value of the random variable.
PDF: A PDF is used to describe the probability distribution of a continuous random variable. It provides the relative likelihood of the random variable taking on different values within a range.

Value Assignments:
PMF: A PMF assigns probabilities to individual values of a discrete random variable. It gives the exact probability of each outcome.
PDF: A PDF gives the relative likelihood (density) of a continuous random variable taking on a particular value. It does not provide the exact probability of any specific value but represents the probability density over an interval.

Probability Calculation:
PMF: For a discrete random variable, the probability of an event is calculated by summing the probabilities of the individual outcomes. The sum of all probabilities in a PMF is equal to 1.
PDF: For a continuous random variable, the probability of an event is calculated by integrating the PDF over the range of values corresponding to that event. The integral of the PDF over the entire range is equal to 1.

Graphical Representation:
PMF: A PMF is typically represented as a probability distribution table or a bar chart, where the values of the random variable are plotted on the x-axis, and their corresponding probabilities are plotted on the y-axis.
PDF: A PDF is typically represented as a smooth curve on a continuous scale, where the x-axis represents the values of the random variable, and the y-axis represents the probability density.

Support:
PMF: The support of a PMF consists of discrete values, and the probability is assigned only to those specific values.
PDF: The support of a PDF consists of an infinite range of values within a continuous interval, and the probability is assigned to different ranges of values rather than individual points.

Normalization:
PMF: A PMF is already normalized by design since the sum of all probabilities must be equal to 1. PDF: A PDF may not be normalized by default. The integral of the PDF needs to be calculated to ensure that the total probability is equal to 1.

At the end we can conclude that the main difference between a PMF and a PDF lies in the type of random variable they describe and the way they assign probabilities. PMFs are used for discrete random variables, assigning probabilities to individual values, while PDFs are used for continuous random variables, describing the relative likelihood of values within a range.


Question7: Explain the Central Limit Theorem (CLT) with example.

Answer: The Central Limit Theorem in Statistics states that as the sample size increases and its variance is finite, then the distribution of the sample mean approaches normal distribution irrespective of the shape of the population distribution. In other words, Central Limit Theorem explains that the sample distribution of the sample mean resembles the normal distribution irrespective of the fact that whether the variables themselves are distributed normally or not. The Central Limit Theorem is often called CLT in abbreviated form.

The central limit theorem posits that the distribution of sample means will invariably conform to a normal distribution provided the sample size is sufficiently large. This holds regardless of the underlying distribution of the population, be it normal, Poisson, binomial, or any alternative distribution.

Central Limit Theorem Formula
Let us assume we have a random variable X. Let σ be its standard deviation and μ is the mean of the random variable. Now as per the Central Limit Theorem, the sample mean X̅ will approximate to the normal distribution which is given as
X̅ ~(N(μ, σ/√n).
The Z-Score of the random variable X̅ is given as Z = (x̄ - μ)/(σ/√n). Here x̄ is the mean of X̅


Example of Central Limit Theorem :
The mean age of cigarette smokers is 35 years. Suppose the standard deviation is 10 years. The sample size is 39. Find the mean and standard deviation of the sample.
Given: μ = 35, σ = 10, n = 39
As per the Central Limit Theorem, the sample mean is equal to the population mean.
Hence, μx̄ = μ = 35 years
Now, σ x̄=σ/√n  = 10/√39
⇒ σx̄ = 1.601 years

We can understand the working of the CLT with an example involving the rolling of a die.
A die has a different number on each side, ranging from 1 to 6. Each number has a one-in-six chance of appearing on a roll. Given the equal likelihood, the dispersion of the numbers that come up from a dice roll is uniform.

using the randint() function to generate the random numbers ranging from 1 to 6.

You will then repeat the process 1000 times. This will give you the result of 1000 sample means. According to CLT, the result of these sample means will be gaussian. The example below shows the resulting distribution of sample means.

```python
import numpy.random as random  # Import random module with alias
import matplotlib.pyplot as plt  # Import pyplot for plotting

# Seed the random number generator (reproducibility)
random.seed(1)

# Calculate the mean of 100 dice rolls 1000 times
means = [random.randint(1, 7, size=100) for i in range(1000)]  # List comprehension
means = [m.mean() for m in means]  # Calculate mean of each roll

# Plot the distribution of sample means
plt.hist(means)
plt.xlabel("Sample Mean (Dice Rolls)")
plt.ylabel("Frequency")
plt.title("Distribution of Sample Means (100 Dice Rolls, 1000 Samples)")
plt.show()
```

Question8:  Compare z-scores and t-scores. When should you use a z-score, and when should a t-score be applied instead?

Answer: - Comparison between Scores vs. T-Scores:

Both z-scores and t-scores are standardized scores used in statistics to measure how many standard deviations a data point is from the mean of a distribution. However, they differ in their underlying assumptions and applications.

Z-Scores
Assumptions: Requires knowledge of the population standard deviation (σ).
Usage: Primarily used when the population standard deviation is known, or when a large sample size (typically n ≥ 30) is available, as the sample standard deviation can be used as a good estimate of the population standard deviation.
Distribution: Follows a standard normal distribution (mean = 0, standard deviation = 1).

T-Scores
Assumptions: Requires knowledge of the sample standard deviation (s).
Usage: Primarily used when the population standard deviation is unknown and the sample size is small (typically n < 30).
Distribution: Follows a t-distribution, which is influenced by the sample size and has wider tails than the standard normal distribution. As the sample size increases, the t-distribution approaches the standard normal distribution.   

In summary:
Z-scores are used when the population standard deviation is known or when a large sample size is available. T-scores are used when the population standard deviation is unknown and the sample size is small. Both z-scores and t-scores are used to standardize data and compare values from different distributions. The choice between z-scores and t-scores depends on the specific conditions of the data and the desired statistical analysis.

Example:

•	If you want to compare the height of a person to the average height of a population, and you know the population standard deviation, you would use a z-score.
•	If you want to compare the average height of a sample of people to the population average, and you don't know the population standard deviation, you would use a t-score.


Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?
Task: Write Python code to calculate the z-score and p-value for the given data.
Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.

```python
import scipy.stats as stats
# Sample mean
sample_mean = 105
# Population mean
population_mean = 100
# Standard deviation
standard_deviation = 15
# Sample size
sample_size = 25
# Calculate the z-score
z_score = (sample_mean - population_mean) / (standard_deviation / sample_size**0.5)
# Calculate the p-value (two-tailed)
p_value = stats.norm.cdf(z_score) * 2
# Significance level
alpha = 0.05
# Check if p-value is less than alpha
if p_value < alpha:
  print("Reject the null hypothesis")
else:
  print("Fail to reject the null hypothesis")
```

Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python. Generate 1,000 samples and plot the distribution. What is the expected mean and variance?
Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.
Objective: Understand the properties of a binomial distribution and verify them through simulation


```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
n = 10  # Number of trials
p = 0.6  # Probability of success

# Generate 1000 samples
samples = np.random.binomial(n, p, size=1000)

# Plot the distribution
plt.hist(samples, bins=n+1, density=True)
plt.title("Binomial Distribution (n=10, p=0.6)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()

# Calculate the expected mean and variance
expected_mean = n * p
expected_variance = n * p * (1 - p)

print("Expected Mean:", expected_mean)
print("Expected Variance:", expected_variance)
```
